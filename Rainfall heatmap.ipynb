```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ── Rainfall data ──────────────────────────────────────────────────────────────

df = pd.read_excel('C:/Users/OEB/Desktop/Book1.xlsx', sheet_name='Sheet2')
 
# Set the first column (Row Labels / Month) as the index
df = df.set_index(df.columns[0])
 
# Ensure all values are numeric
df = df.apply(pd.to_numeric, errors='coerce')

# ── Plot ───────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 7))

sns.heatmap(
    df,
    ax=ax,
    annot=True,           # show values in each cell
    fmt="d",              # integer format
    cmap="YlGnBu",        # yellow → green → blue (dry → wet)
    linewidths=0.4,
    linecolor="white",
    cbar_kws={"label": "Rainfall (mm)", "shrink": 0.8, "pad": 0.02},
    annot_kws={"size": 8},
)

ax.set_title("Monthly Rainfall Heatmap (2010-2023)", fontsize=11, fontweight="bold", pad=10)
ax.set_xlabel("Year", fontsize=11, labelpad=8)
ax.set_ylabel("Month", fontsize=11)
ax.tick_params(axis="x", length=0)
ax.tick_params(axis="y", rotation=0, length=0)

plt.tight_layout()
plt.show()
# plt.savefig("rainfall_heatmap.jpeg", dpi=150, bbox_inches="tight")
# print("Saved → rainfall_heatmap.png")
```
